```python
class PositionNode:

    def __init__(self, x, y, val):
        self.x = x
        self.y = y
        self.val = val
        self.connected_nodes = [] # max 2 connected nodes
        self.connected = False
        self.next_node = None
        
    @property
    def position(self):
        return (self.x, self.y)
    
    def __repr__(self):
        return f"{self.val}"
    
    # @property
    # def direction_1(self):

    
    # @property
    # def direction_2(self):

    

    def connect_me(self, grid, referred_by=None):

        if self.connected == False:
            self.connected = True
            print(f"connecting {self.position}")

            if self.val == "|":
                # north south
                self.connected_nodes = [grid[self.x-1][self.y],
                                        grid[self.x+1][self.y]]

            if self.val == "-":
                # west east
                self.connected_nodes = [grid[self.x][self.y-1],
                                        grid[self.x][self.y+1]]

            if self.val == "L":
                # north east
                self.connected_nodes = [grid[self.x-1][self.y],
                                        grid[self.x][self.y+1]]

            if self.val == "J":
                # north and west
                self.connected_nodes = [grid[self.x-1][self.y],
                                        grid[self.x][self.y-1]]

            if self.val == "7":
            #  south and west
                self.connected_nodes = [grid[self.x+1][self.y],
                                        grid[self.x][self.y-1]]

            if self.val == "F":
            # south and east
                self.connected_nodes = [grid[self.x+1][self.y],
                                        grid[self.x][self.y+1]]

            if self.val == ".":
                # no connection
                pass

            if self.val == "S":
                # it's an S to the right, and a J to the bottom
                # east and south == F
                self.connected_nodes = [grid[self.x+1][self.y],
                                        grid[self.x][self.y+1]]

                self.connected_nodes[0].connect_me(grid, self)
                
                # starting point
            
            self.next_node = [x for x in self.connected_nodes if x != referred_by]
            # print(len(self.next_node))
        # return self.next_node
     
            if len(self.next_node) == 1:
                self.next_node[0].connect_me(grid, self)
        
```
