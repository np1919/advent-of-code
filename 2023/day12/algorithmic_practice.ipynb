```python
def can_construct(target:str, wordbank:list):
    print(target)

    if target == "":
        return True

    for possibles in wordbank:
        if target.startswith(possibles):
            result = can_construct(target[len(possibles):], wordbank)
            if result == True:
                return True

    return False

can_construct("aaaaaaaab", ['a'])
```
