```python
# day 3 part 2
with open('puzzle_input.txt', 'r') as f:
    data = [x.strip('\n').strip(' ') for x in f.readlines()]

# import string
import re
# from collections import defaultdict


    # all symbols except for `.`
symbols = ['*']
# symbols.remove('.')


class Node:
    def __init__(self, rownum, colnum, value):
        self.rownum = rownum
        self.colnum = colnum
        self.value = value
        self.connected = False
        self.is_gear = False
        self.connected_integers = []

        pass
    
    def __repr__(self):
        return f"{self.value} at {self.rownum, self.colnum}"


all_nodes = []

for rownum, row in enumerate(data):
    thisrow = []
    #current = None
    for colnum, column in enumerate(row):
        current = Node(rownum, colnum, data[rownum][colnum])
        thisrow.append(current)
    all_nodes.append(thisrow)

all_gears = []

for rownum, row in enumerate(all_nodes):
    for colnum, cell in enumerate(row):
        if cell.value in symbols:
            cell.connected=True
            cell.is_gear=True
            all_gears.append(cell)


            ## add the connected indexes
            for i in range(cell.rownum-1, cell.rownum+2):
                for j in range(cell.colnum-1, cell.colnum+2):
                    try:
                        all_nodes[i][j].connected = True
                    except:
                        print('error') # no symbols are out of bounds?? crazy
                        print('out of bounds for row {i} column {j}')
# all_nodes.__len__(), [x.__len__() for x in all_nodes]

# for row in all_nodes:
#     print([True if x.connected == True else None for x in row])

for cell in all_gears:
    connected_to_gear = []
    # print(f"searching for connected elements to gear at {cell}")

    for i in (-1, 0, +1):
        row = data[cell.rownum+i]
        matches = re.finditer('\d+', row)
        # print(f"matches for row {cell.rownum+i}")
        for m in matches:
            # print(m)
            for index in list(range(*m.span())):
                if index in range(cell.colnum-1, cell.colnum+2):
                    connected_to_gear.append(m.group())
                    break
                
            
        
    cell.connected_integers = connected_to_gear
    # print(cell, connected_to_gear)




total_value = 0
for gear in all_gears:
    if len(gear.connected_integers) == 2:
        total_value += int(gear.connected_integers[0]) * int(gear.connected_integers[1])

total_value

#91031374 is correct
```
