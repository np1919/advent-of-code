```python
import numpy as np

def scan_vertical(entry):
    active = []
    for line in entry:
        active.append(line)
        print(active[:len(active)//2],'*',active[len(active)//2:])
        if active[:len(active)//2] == active[len(active)//2:]:
            print('found a reflection')

    active = []
    for line in entry[::-1]:
        active.append(line)
        print(active[:len(active)//2],'*',active[len(active)//2:])
        if active[:len(active)//2] == active[len(active)//2:]:
            print('found a reflection')

def scan_horizontal(entry):
    active = []
    for idx, _ in enumerate(entry[0]):
        line = "".join([x[idx] for x in entry])
        active.append(line)
        print(active[:len(active)//2],'*',active[len(active)//2:])
        if active[:len(active)//2] == active[len(active)//2:]:
            print('found a reflection')
    active = []
    for idx in range(len(entry[0]), 0):
        line = "".join([x[idx] for x in entry])
        active.append(line)
        print(active[:len(active)//2],'*',active[len(active)//2:])
        if active[:len(active)//2] == active[len(active)//2:]:
            print('found a reflection')
```
