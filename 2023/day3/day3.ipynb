```python
import string 
from collections import defaultdict
import re

# all symbols except for `.`
symbols = [x for x in string.punctuation]
symbols.remove('.')

# dictionary of {rownumber : {int : range(start_index, end_index)}}
integer_map = defaultdict(dict)

# dictionary of {rownumber : [index1, index2...]}
symbol_map = dict()

for rownumber, line in enumerate(data):
    # symbol indexes
    symbol_indexes = []
    for idx, char in enumerate(line):
        if char in symbols:
            symbol_indexes.append(idx)
    symbol_map[rownumber] = symbol_indexes

    # numbers and index ranges
    matches = re.finditer('\d+', line)
    match_values = re.findall('\d+', line)
    integer_map[rownumber] = dict(zip(match_values, [(m.span()[0], m.span()[1]) for m in re.finditer('\d+', line)])) 
```

```python
symbol_indices = defaultdict(set)

# process the 'connected' values for each row
# we will extend the range for each integer and keep the symbols as single values
for row in symbol_map:
    for i in row-1, row, row+1:
        try:
            [symbol_indices[row].add(x) for x in symbol_map[i]]
        except KeyError:
            pass
    # print(row, sorted(set(symbol_indices)))
```

```python

for row in symbol_map:
    assert all([y in symbol_indices[row] for y in symbol_map[row]])
```

```python
for row in integer_map:
    for integer in integer_map[row]:
        assert len(integer_map[row][integer]) == 2
```

```python

# connected parts
connected_parts = defaultdict(list)
not_connected = defaultdict(list)

# for each row
for row in integer_map:
    # of integers and their index ranges
    for integer in integer_map[row]:
        if any(x in range(integer_map[row][integer][0]-1, integer_map[row][integer][1]+1) for x in symbol_indices[row]):
            # print(f'match found for {integer}')
            # print('match found')
            connected_parts[row].append(int(integer))
            # break
        else:
            not_connected[row].append(int(integer))
        
```
